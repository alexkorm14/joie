```python
def unit_df(path_train, path_test):
  df_train = pd.read_csv(path_train, sep='	', header=None)
  df_test = pd.read_csv(path_test, sep='	', header=None)
  df_unit = pd.concat([df_train, df_test])
  return df_unit

unit_df('data/yago/yago_InsType_train.txt', 'data/yago/yago_InsType_test.txt').to_csv('data/yago/yago_InsType.txt', sep='	', header=None)

unit_df('data/yago/yago_insnet_train.txt', 'data/yago/yago_insnet_test.txt').to_csv('data/yago/yago_insnet.txt', sep='	', header=None)

unit_df('data/yago/yago_ontonet_train.txt', 'data/yago/yago_ontonet_test.txt').to_csv('data/yago/yago_ontonet.txt', sep='	', header=None)
```
